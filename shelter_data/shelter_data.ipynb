```python
import csv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
pd.options.display.float_format = '{:,.2f}'.format
```

```python
Sh09_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2009.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh10_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2010.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh11_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2011.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh12_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2012.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh13_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2013.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh14_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2014.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh15_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2015.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh16_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2016.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh17_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2017.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
Sh18_df = pd.read_csv("2007-2018-HIC-Counts-by-State_2018.csv", usecols=[0,1,4], header=1, names=["State","Total Beds","Participation"])
```

```python
Sh09_df['Year'] = '2009'
Sh10_df['Year'] = '2010'
Sh11_df['Year'] = '2011'
Sh12_df['Year'] = '2012'
Sh13_df['Year'] = '2013'
Sh14_df['Year'] = '2014'
Sh15_df['Year'] = '2015'
Sh16_df['Year'] = '2016'
Sh17_df['Year'] = '2017'
Sh18_df['Year'] = '2018'
```

```python
dfs=[Sh09_df,Sh10_df,Sh11_df,Sh12_df,Sh13_df,Sh14_df,Sh15_df,Sh16_df,Sh17_df,Sh18_df]
shelter_data_all = pd.concat(dfs, axis=0, join='outer', join_axes=None, ignore_index=False,
          keys=None, levels=None, names=None, verify_integrity=False,
          copy=True)
```

```python
shelter_data_all=shelter_data_all[shelter_data_all.State != 'Total']
shelter_data_all
```

```python
shelter_data_all = shelter_data_all[['State','Year','Total Beds','Participation']]
shelter_data_all = shelter_data_all.sort_values(by=['State','Year'])

shelter_data_all
```

```python
shelter_data_merged=shelter_data_all
shelter_data_merged.to_csv("Shelters by State -2009-18.csv", mode='a',index=False)
```
